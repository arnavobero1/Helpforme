# Electric Motor Speed Prediction

### Purpose: 
#### The sole purpose of choosing and analyzing this dataset is to understand how well the electric cars work and how efficient they are in time like this when they are actually just starting to boom.


- ambient = Ambient temperature as measured by a thermal sensor located closely to the stator.
- coolant = Coolant temperature. The motor is water cooled. Measurement is taken at outflow.
- u_d = Voltage d-component
- u_q = Voltage q-component
- motor_speed = Motor speed
- torque= Torque induced by current.
- i_d = Current d-component
- i_q = Current q-component
- pm = Permanent Magnet surface temperature representing the rotor temperature. This was measured with an infrared thermography unit.
- stator_yoke = Stator yoke temperature measured with a thermal sensor.
- stator_tooth = Stator tooth temperature measured with a thermal sensor.
- stator_winding = Stator winding temperature measured with a thermal sensor.
- profile_id = Each measurement session has a unique ID. Make sure not to try to estimate from one session onto the other as they are strongly independent.


##### * key note: Recommended split - 75% Training and 25% Testing

### Step 3: Data Preprocessing and Correlations

```python
df.drop('profile_id',axis=1,inplace=True)
```

```python
df.skew()
```

#### Skewness of ambient and coolant readings are a little over normal distribution's requirements

```python
df.kurtosis()
```

- Kurtosis of ambient, i_q, torque is Leptokurtic. All the other features show Playtkurtic nature.

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
df.skew()
```

- Improved skewness due to outlier removal

```python
plt.figure(figsize = (10,8))
sns.heatmap(df.corr(), annot=True, linewidths=.5,square=True,cmap = 'viridis');
```

### Step 4: Feature Selection using Fisher Feature Selection

```python
row,col = df.shape
def fisher(parameter):
    
    u1 = parameter[:499035]
    u2 = parameter[499035:]
    mu1 = u1.mean()
    mu2 = u2.mean()
    var1 = u1.var()
    var2 = u2.var()
    return mu1,mu2,var1,var2

feature_list = []
for i in range(col):
    mean1,mean2,var1,var2 = fisher(df.iloc[:,i])
    feature_list.append([mean1,mean2,var1,var2])
```

### \begin{align}
Score = \frac{\mu_1 -\mu_2}{\sigma_1^2+\sigma_2^2} 
\end{align}

```python
#Checking the score in descending order
list2 = []
i = 0
for elements in feature_list:
    list2.append([(elements[0]-elements[1])/(elements[2]+elements[3]),'{}'.format(df.columns[i])])
    i += 1
    
list2.sort(reverse = True)
list2
```

### Step 7: Preparing classification models
# Classification

- We will assume speed from -3 to -1 as "Slow"
- Speed of -1 to 1 as "Moderate"
- Speed of 1 to 3 as "Fast"

#### Data Preprocessing for Classification

```python
def speed_classifier(x):
    if (x < -1):
        return 'Slow'
    elif (x >= -1) and (x <= 1):
        return 'Moderate'
    else:
        return 'Fast'
```

```python
df['motor_speed_new'] = df['motor_speed'].apply(speed_classifier)
```
