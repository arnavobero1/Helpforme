```python
#Importing the libraries
import numpy as np
import pandas as pd

#Initializing the variables and dataframe
a = np.random.randn(50)
b = np.random.randn(50)
c = np.random.randn(50)
d = np.random.randn(50)
e = np.random.randn(50)
f = np.random.randn(50)
g = np.random.randn(50)
h = np.random.randn(50)
i = np.random.randn(50)
j = np.random.randn(50)
k = np.random.randn(50)
l = np.random.randn(50)
m = np.random.randn(50)
n = np.random.randn(50)
o = np.random.randn(50)
p = np.random.randn(50)
q = np.random.randn(50)
r = np.random.randn(50)
s = np.random.randn(50)
t = np.random.randn(50)

df = pd.DataFrame([a,b,c,d,e,f,g,h,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t],index=['a','b','c','d','e','f','g','h','e','f','g',
                                                                           'h','i','j','k','l','m','n','o','p','q','r',
                                                                           's','t']).transpose()
```

```python
#Checking our dataset
df.head()
```

```python
#Creating the fisher filter feature selection then appending on list
def fisher(param):
    u1 = param[:25]
    u2 = param[25:]
    u1m = u1.mean()
    u2m = u2.mean()
    var1 = u1.var()
    var2 = u2.var()
    return u1m,u2m,var1,var2
feature_list = []
for i in range(50):
    mean1,mean2,var1,var2 = fisher(df.iloc[:,i])
    feature_list.append([mean1,mean2,var1,var2])
```

```python
#Creating feature name
import string
string.ascii_lowercase
'abcdefghijklmnopqrst'
feature = list(string.ascii_lowercase)
```

### \begin{align}
Score = \frac{\mu_1 -\mu_2}{\sigma_1^2+\sigma_2^2} 
\end{align}

```python
#Checking the score in descending order
list2 = []
i = 0
for elements in feature_list:
    list2.append([(elements[0]-elements[1])/(elements[2]+elements[3]),'{}'.format(feature[i])])
    i += 1
    
list2.sort(reverse = True)
```
