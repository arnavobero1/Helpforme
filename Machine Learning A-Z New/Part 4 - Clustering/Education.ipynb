## Preparing dataset for clustering:

```python
X = df.copy()
```

```python
X.head()
```

```python
#droping the unwanted columns
X.drop(['gender','NationalITy','PlaceofBirth','StageID','GradeID','Topic','Semester','Relation','ParentAnsweringSurvey','ParentschoolSatisfaction','StudentAbsenceDays'], axis=1, inplace=True)
```

```python
X.head()
```

```python
# normalize the values in x_data
X=(X-np.min(X))/(np.max(X)-np.min(X))
```

##### Extracting the original labels provided with the dataset to check how accurate are the clustering methods.

```python
Z = df.copy()
```

```python
#droping the unwanted columns
Z.drop(['NationalITy','PlaceofBirth','StageID','GradeID','Topic','Semester','Relation','raisedhands','VisITedResources','AnnouncementsView','Discussion','ParentAnsweringSurvey','ParentschoolSatisfaction','StudentAbsenceDays'], axis=1, inplace=True)
```

```python
Z.head()
```

```python
#Function to clean column
def sex(param):
    if param == 'M':
        return 1
    else:
        return 0
Z['gender'] = Z['gender'].apply(sex)
```

```python
Z.head()
```

# K-MEANS CLUSTERING:

## ELBOW METHOD:

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=300, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### We can see that number of clusters = 2 or 3 as indicated by the elobow method. 

## SILHOUETTE COEFFICIENT:

```python
from sklearn.metrics import silhouette_score
silhouette_coefficients = []
# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=300, n_init=10)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.title('Silhouette Coefficients')
plt.xlabel('Number of clusters')
plt.ylabel('silhouette_coefficients')
plt.show()
```

#### From the graph we can see that number of clusters = 2 or 10 have highest values of silhoutte coefficient.

# HIERARCHIAL CLUSTERING:

## DENDOGRAM:

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```
