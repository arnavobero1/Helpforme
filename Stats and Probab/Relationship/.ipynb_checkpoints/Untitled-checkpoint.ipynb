```python
import brfss
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import random
import thinkplot
import thinkstats2
```

```python
df = brfss.ReadBrfss(nrows=None)
```

```python
def SampleRows(df,nrows,replace=False):
    indices=np.random.choice(df.index,nrows,replace=replace)
    sample=df.loc[indices]
    return sample
```

```python
sample=SampleRows(df,5000)
heights,weights=sample.htm3,sample.wtkg2
```

```python
plt.figure(figsize=(16,9))
thinkplot.Scatter(heights,weights,alpha=1)
thinkplot.Config(xlabel='Height(cm)',ylabel='Weight(kg)',axis=[140,210,20,200],legend=False)
```

```python
def Jitter(values,jitter=0.5):
    n=len(values)
    #return np.random.uniform(-jitter,+jitter,n)+values
    return np.random.normal(0,jitter,n)+values
```

```python
heights=Jitter(heights,1.4)
weights=Jitter(weights,0.5)
```

```python
plt.figure(figsize=(16,9))
thinkplot.Scatter(heights,weights,alpha=1)
thinkplot.Config(xlabel='Height(cm)',ylabel='Weight(kg)',axis=[140,210,20,200],legend=False)
```

```python
plt.figure(figsize=(16,9))
thinkplot.Scatter(heights,weights,alpha=0.1,s=10)
thinkplot.Config(xlabel='Height(cm)',ylabel='Weight(kg)',axis=[140,210,20,200],legend=False)
```

```python
plt.figure(figsize=(16,9))
thinkplot.HexBin(heights,weights)
thinkplot.Config(xlabel='Height(cm)',ylabel='Weight(kg)',axis=[140,210,20,200],legend=False)
```

```python
cleaned=df.dropna(subset=['htm3','wtkg2'])
```

```python
bins=np.arange(135,210,5)
indices=np.digitize(cleaned.htm3,bins)
groups=cleaned.groupby(indices)
```

```python
for i,group in groups:
    print(i,len(group))
```

```python
mean_heights=[group.htm3.mean() for i, group in groups]
cdfs=[thinkstats2.Cdf(group.wtkg2) for i, group in groups]
```

```python
mean_heights
```

```python
for percent in [75,50,25]:
    weight_percentiles=[cdf.Percentile(percent) for cdf in cdfs]
    label='%dth' % percent
    thinkplot.Plot(mean_heights,weight_percentiles,label=label)
thinkplot.Config(xlabel='Height(cm)',ylabel='Weight(kg)',axis=[135,210,20,130],legend=True)
```

```python
def Cov(xs,ys,meanx=None,meany=None):
    xs=np.asarray(xs)
    ys=np.asarray(ys)
    
    if meanx is None:
        meanx=np.mean(xs)
    if meany is None:
        meany = np.mean(ys)
        
    cov=np.dot(xs-meanx,ys-meany)/len(xs)
    return cov
```

```python
heights,weights=cleaned.htm3,cleaned.wtkg2
Cov(heights,weights)
```

```python
def Corr(xs,ys):
    xs=np.asarray(xs)
    ys=np.asarray(ys)
    
    meanx,varx=thinkstats2.MeanVar(xs)
    meany,vary=thinkstats2,MeanVar(ys)
    
    corr=Cov(xs,ys,meanx,meany)/np.sqrt(varx*vary)
    return corr
```
